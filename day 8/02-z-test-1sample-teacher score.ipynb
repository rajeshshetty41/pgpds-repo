------------------------------------
- Qs : A teacher claims that the mean score of students in his class is greater than 82 with a standard deviation of 20. If a sample of 81 students was selected with a mean score of 90 then check if there is enough evidence to support this claim at a 0.05 significance level.

-------------------------------

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Given data
population_mean = 82
sample_mean     = 90
population_std  = 20
sample_size     = 81

alpha = 0.05
```

#### State the Null and Alternate Hypothesis

- Null Hypothesis (H0): The mean score of students is less than or equal to 82.
    - ð»0:ðœ‡â‰¤82

- Alternate Hypothesis (H1): The mean score of students is greater than 82.
    - ð»1:ðœ‡>82


#### Describe the Type of Z-test
We will use a one-sample Z-test for this hypothesis because:

The population standard deviation is known (20).
The sample size is large (81).

#### Compute the Test Statistic
The test statistic for a one-sample Z-test is calculated using the formula:



```python
# Calculate the Z-test statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_stat
```

```python
# Critical value for one-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)
z_critical
```

```python
# Calculate the p-value
p_value = 1 - stats.norm.cdf(z_stat)
p_value
```

```python
# Print the results
print(f"Z-test Statistic: {z_stat}")
print(f"Critical Value: {z_critical}")
print(f"P-value: {p_value}")
```

```python
# Plot the results
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Distribution')

# Shade the rejection region
plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region')

# Plot the test statistic
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Test Statistic (Z = {z_stat:.2f})')

# Add labels and legend
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.title('One-Sample Z-Test')
plt.legend()

# Show the plot
plt.show()
```

```python
# Conclusion
if z_stat > z_critical:
    print("Reject the null hypothesis: There is sufficient evidence to support the teacher's claim.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to support the teacher's claim.")
```
