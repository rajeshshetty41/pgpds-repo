#### assumptions for t-test
-------------------

- Normality

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, probplot
```

```python
# 1. Generate Dummy Data
np.random.seed(0)  # for reproducibility
data = np.random.randn(40)
```

**Histogram and KDE Plot**: 

- This combined plot provides a visualization of the data's distribution. 
- The histogram shows frequency counts in bins, while the KDE gives an estimated probability density function. 
- If the data is normal, the KDE should roughly resemble the shape of a bell curve.

```python
# 2. Histogram and KDE Plot
plt.figure(figsize=(6, 4))
sns.histplot(data, kde=True)
plt.title('Histogram and KDE Plot')
plt.xlabel('Value')
plt.ylabel('Frequency');
```

**Q-Q Plot**: 

- In this plot, the quantiles of the data are plotted against the quantiles of a normal distribution.
- If the data is normal, the points should lie on the 45-degree reference line. Deviations from this line suggest deviations from normality.

```python
# 4. Q-Q Plot
plt.figure(figsize=(10, 5))
probplot(data, plot=plt)
plt.title('Q-Q Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles');
```

**Test of Normality**: 
    
- The Shapiro-Wilk test tests the null hypothesis that the data was drawn from a normal distribution. A low p-value (typically below 0.05) indicates that the data may not be normally distributed.

- H0: The data is normally distributed.
- H1: The data is not normally distributed.

```python
# 3. Test of Normality using Shapiro-Wilk
stat, p_value = shapiro(data)

print(f'Shapiro-Wilk Statistic: {stat:.4f}')
print(f'P-Value: {p_value:.4f}')
if p_value > 0.05:
    print('Data appears to be normally distributed (fail to reject H0)')
else:
    print('Data may not be normally distributed (reject H0)')
```
