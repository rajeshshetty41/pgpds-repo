```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Customer data: age and income
customer_data = np.array([
    [25, 50000],
    [30, 75000],
    [35, 60000],
    [40, 80000],
    [45, 90000]
])
```

```python
# Select three customers
selected_customers = customer_data[:3]
```

```python
# Plot the customer data
plt.figure(figsize=(8, 6))

# Plot all customers' data
plt.scatter(customer_data[:, 0], customer_data[:, 1], color='blue', label='All Customers')

# Highlight the selected customers
plt.scatter(selected_customers[:, 0], selected_customers[:, 1], color='red', label='Selected Customers')

# Add labels for the selected customers
for i, (age, income) in enumerate(selected_customers):
    plt.annotate(f'Customer {i+1}\n({age}, {income})', (age, income), textcoords="offset points", xytext=(-15, 7), ha='center', fontsize=9, color='red')

# Plot vectors from origin to selected customers
origin = [0, 0]
vectors = selected_customers - origin

for i, (age, income) in enumerate(selected_customers):
    plt.quiver(*origin, vectors[i, 0], vectors[i, 1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
    plt.annotate(f'Vector {i+1}', (vectors[i, 0]/2, vectors[i, 1]/2), textcoords="offset points", xytext=(-10, 7), ha='center', fontsize=9, color='green')

# Add labels and legend
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Customer Information Data with Vectors')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

#### scalars

```python
# Define two scalar values
scalar1 = 5
scalar2 = 3
```

```python
# Plot a number line
plt.figure(figsize=(8, 1))
plt.axhline(y=0, color='black', linewidth=2)

# Mark the scalar values
plt.scatter(scalar1, 0, color='blue', label='Scalar 1')
plt.scatter(scalar2, 0, color='red', label='Scalar 2')

# Add labels for the scalar values
plt.annotate(f'Scalar 1: {scalar1}', (scalar1, 0), textcoords="offset points", xytext=(-20, 10), ha='center', fontsize=9, color='blue')
plt.annotate(f'Scalar 2: {scalar2}', (scalar2, 0), textcoords="offset points", xytext=(20, 10), ha='center', fontsize=9, color='red')

# Set plot limits and labels
plt.xlim(min(scalar1, scalar2) - 2, max(scalar1, scalar2) + 2)
plt.ylim(-5, 10)
plt.xticks([])  # Hide x-axis labels
plt.yticks([])  # Hide y-axis labels
plt.title('Scalar Values on Number Line')
plt.legend()

# Show the plot
plt.show()
```

#### vector spaces

- A vector space (also called a linear space) is a collection of objects called vectors

```python
# Define vectors in R^2
vectors = np.array([[2, 3]])

# Create a figure and set axis limits
plt.figure(figsize=(8, 6))
plt.xlim(0, 4)
plt.ylim(0, 4)

# Plot the vectors
plt.quiver([0], [0], vectors[:, 0], vectors[:, 1], angles='xy', scale_units='xy', scale=1, color=['blue'], label=['Vector 1'])

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Space Example: Euclidean Space (R^2)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define vectors in R^3
vectors = np.array([[1, 2, 3], 
                    [4, 0, 2], 
                    [3, 2, 1]])
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, vectors[:, 0], vectors[:, 1], vectors[:, 2], color=['blue', 'red', 'green'], label=['Vector 1', 'Vector 2', 'Vector 3'])

# Set plot limits
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 4])

# Add labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Space Example: Euclidean Space (R^3)')
ax.legend()

# Show the plot
plt.show()
```

#### magnitude

- In linear algebra, the norm of a vector is a mathematical concept that quantifies the length or size of the vector in a vector space. 

- The norm provides a measure of the `magnitude` of the vector, which is often used to express the concept of distance or size in various contexts. 

- Different types of norms exist, each with its own properties and applications. 

- The most common norm used is the Euclidean norm (L2 norm), but other norms like the L1 norm (Manhattan norm) and the L norm (maximum norm) are also frequently used.

**L2 Norm**

The formula for calculating the L2 norm (also known as the Euclidean norm) 

of a vector (x1, x2, ..xn) is 



```python
# Example vector
vector = np.array([3, 4])

# Calculate Euclidean norm (L2 norm)
euclidean_norm = np.linalg.norm(vector)
print("Euclidean Norm of the Vector:", euclidean_norm)
```

```python
# Example vector
vector = np.array([3, 4])

# Calculate L2 norm (Euclidean norm)
l2_norm = np.sqrt(np.sum(vector**2))
print("L2 Norm (Euclidean Norm) of the Vector:", l2_norm)
```

```python
# Example data: [area, price]
house1 = np.array([1500, 250000])
house2 = np.array([1800, 320000])
house3 = np.array([1200, 200000])

# Calculate magnitudes
magnitude1 = np.linalg.norm(house1)
magnitude2 = np.linalg.norm(house2)
magnitude3 = np.linalg.norm(house3)

print("Magnitude of House 1:", magnitude1)
print("Magnitude of House 2:", magnitude2)
print("Magnitude of House 3:", magnitude3)
```

#### Directions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example feature vectors (number of bedrooms, square footage)
house_a = np.array([3, 1500])
house_b = np.array([4, 1800])
house_c = np.array([2, 1200])
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter([house_a[0], house_b[0], house_c[0]],
            [house_a[1], house_b[1], house_c[1]], color='blue')

# Add arrows to represent feature vector directions
plt.quiver(0, 0, house_a[0], house_a[1], angles='xy', scale_units='xy', scale=1, color='red', label='House A')
plt.quiver(0, 0, house_b[0], house_b[1], angles='xy', scale_units='xy', scale=1, color='green', label='House B')
plt.quiver(0, 0, house_c[0], house_c[1], angles='xy', scale_units='xy', scale=1, color='purple', label='House C')

# Add magnitude labels
plt.text(house_a[0] + 0.2, house_a[1], f'Magnitude: {np.linalg.norm(house_a):.2f}', color='red')
plt.text(house_b[0] + 0.2, house_b[1], f'Magnitude: {np.linalg.norm(house_b):.2f}', color='green')
plt.text(house_c[0] + 0.2, house_c[1], f'Magnitude: {np.linalg.norm(house_c):.2f}', color='purple')

plt.xlabel('Number of Bedrooms')
plt.ylabel('Square Footage')
plt.title('Feature Vector Directions with Magnitude')
plt.legend(loc='upper left')
plt.grid()
plt.xlim(0, 5)
plt.ylim(0, 2000);
```
