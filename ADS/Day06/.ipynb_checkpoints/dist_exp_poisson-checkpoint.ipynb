```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns    # plot
```

```python
# Set a common seed for reproducibility
np.random.seed(42)
```

#### 1. ND

```python
data_normal = np.random.normal(loc  = 0.0, 
                               scale= 1.0, 
                               size = 1000)
```

```python
plt.figure(figsize=(5, 5))

sns.histplot(data_normal, kde=True)

plt.title('Normal Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_normal.min(), data_normal.max()))

plt.xlabel('Values')
plt.ylabel('Frequency');
```

```python
# 2. Uniform Distribution

data_uniform = np.random.uniform(low=0.0, high=1.0, size=1000)

plt.figure(figsize=(5, 5))
sns.histplot(data_uniform, kde=True)
plt.title('Uniform Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_uniform.min(), data_uniform.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Exponential Distribution
data_exponential = np.random.exponential(scale=1.0, size=1000)

plt.figure(figsize=(5, 5))

sns.histplot(data_exponential, kde=True)

plt.title('Exponential Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_exponential.min(), data_exponential.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Key Points about the Exponential Distribution

1. **Nature of Exponential Distribution:**
   - The exponential distribution is a continuous probability distribution used to `model the time or space between events in a Poisson process`, where events occur continuously and independently at a constant rate.

2. **Shape of the Distribution:**
   - The distribution is typically right-skewed, meaning most values cluster near the lower end of the range, with a long tail extending towards higher values. This reflects the nature of the process, where many events happen quickly, but some take a longer time to occur.

3. **Parameter of the Distribution ( - Lambda):**
   - The exponential distribution is characterized by the rate parameter $\lambda$, which is the reciprocal of the mean. If $\lambda$ is large, events happen quickly (small gaps between events), and if $\lambda$ is small, the events occur more slowly (larger gaps between events).
   - In the code, the default rate parameter $\lambda = 1 $ is used by `np.random.exponential()`, resulting in a mean of 1.

4. **Applications:**
   - The exponential distribution is commonly used to model waiting times, such as the time between arrivals at a service point (e.g., customers arriving at a bank, radioactive decay events, or time until the next earthquake).

5. **Memoryless Property:**
   - A key property of the exponential distribution is the "memoryless" property, meaning the probability of an event occurring in the future is independent of how much time has already passed.

6. **Range and Interpretation:**
   - The generated samples in the code are drawn from this distribution, and the range of the data is from `data_exponential.min()` to `data_exponential.max()`. The histogram and KDE plot illustrate the distribution, where most of the data points are near zero, with the probability of larger values decreasing exponentially.

7. **Visualization:**
   - The histogram with a kernel density estimate (KDE) plot provides a visual representation of the distribution of values. The sharp peak near zero and the gradual decline towards the right are characteristic of the exponential distribution.

This distribution is fundamental in various fields such as queuing theory, reliability engineering, and survival analysis.

#### Poisson distribution

```python
# Expected number of events occurring in a fixed-time interval,
# must be >= 0. A sequence must be broadcastable over the requested size.

data_poisson = np.random.poisson(lam=2, size=1000)

plt.figure(figsize=(5, 5))
sns.histplot(data_poisson, kde=True)
plt.title('Poisson Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_poisson.min(), data_poisson.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Key Points about the Poisson Distribution

1. **Nature of Poisson Distribution:**
   - The Poisson distribution is a discrete probability distribution that models the number of events occurring within a fixed interval of time or space, where these events happen with a constant mean rate and independently of the time since the last event.

2. **Parameter of the Distribution ( - Lambda):**
   - The Poisson distribution is characterized by the parameter \( \lambda \) (lambda), which represents the average number of events in the given interval.
   - In the code, `lam=3` means the average rate of events is 3 per interval.

3. **Shape of the Distribution:**
   - The distribution is `typically right-skewed`, especially for smaller values of \( \lambda \). As \( \lambda \) increases, the distribution becomes more symmetric and approaches a normal distribution.
   - The distribution's mean and variance are both equal to \( \lambda \).

4. **Applications:**
   - The Poisson distribution is often used to model random events that happen independently and at a constant average rate, such as:
     - The number of phone calls received by a call center in an hour.
     - The number of decay events per unit time from a radioactive source.
     - The number of customers arriving at a store within a certain period.

5. **Range and Interpretation:**
   - The generated samples in the code are drawn from this distribution, with values ranging from `data_poisson.min()` to `data_poisson.max()`. The histogram displays the frequency of different outcomes (number of events) across the 1000 samples.

6. **Mean and Variance:**
   - Both the mean and variance of the Poisson distribution are equal to \( \lambda \). For this example, with \( \lambda = 3 \), the mean and variance are both 3.

7. **Visualization:**
   - The histogram provides a visual representation of the distribution of the number of events across the 1000 samples. The plot illustrates how frequently different event counts (e.g., 0, 1, 2, etc.) occur within the interval. The use of a KDE plot here is not typical, as its more suitable for continuous distributions, but it can still provide a smooth visualization of the data.

The Poisson distribution is crucial in fields like queueing theory, risk analysis, and various domains where the modeling of rare events is required.

```python
# List of lambda values
lambdas = [2, 4, 6, 8, 10, 12, 14, 500]

# Create subplots
plt.figure(figsize=(16, 8))

for i, lam in enumerate(lambdas):
    plt.subplot(2, 4, i+1)
    data_poisson = np.random.poisson(lam=lam, size=1000)
    sns.histplot(data_poisson, kde=False, bins=30, color='blue')
    plt.title(f'Poisson Distribution\n = {lam}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Example: Call Center

**Scenario:**

A call center receives phone calls from customers. We want to model two aspects of this scenario:

- **Time Between Calls (Exponential Distribution):**
  We want to model the time we expect to wait between receiving consecutive phone calls.
  
- **Number of Calls Received in an Hour (Poisson Distribution):**
  We want to model the number of phone calls received in a fixed time period, such as an hour.

#### 1. Exponential Distribution: Time Between Calls

**Context:**

Suppose the average rate of receiving phone calls is 5 calls per hour. This rate is denoted by \( \lambda \) in the Exponential distribution context.

**Parameter:**

\( \lambda = 5 \) calls per hour (rate of calls).

**Explanation:**

The Exponential distribution models the time between successive phone calls. If the rate of calls is 5 per hour, the average time between calls is $ \frac{1}{\lambda} $ hours, which is $ \frac{1}{5} $ hours (or 12 minutes).

```python
# Rate of calls
lambda_exponential = 5

# Generate data
data_exponential = np.random.exponential(scale=1/lambda_exponential, size=1000)

# Convert hours to minutes
data_exponential_minutes = data_exponential * 60

# Plot
plt.figure(figsize=(8, 4))

sns.histplot(data_exponential_minutes, kde=True, bins=30, color='blue')

plt.title(f'Exponential Distribution - Time Between Calls\n = {lambda_exponential}, Mean Time = {60 / lambda_exponential:.2f} minutes')
plt.xlabel('Time Between Calls (minutes)')
plt.ylabel('Frequency')
plt.show()
```

#### 2. Poisson Distribution: Number of Calls in an Hour

**Context**:
We want to model the number of phone calls received in one hour.

**Parameter**
 = 5 (average number of calls per hour)

**Explanation**
- The Poisson distribution models the number of calls received in an hour.
- If the average rate is 5 calls per hour, then the number of calls received follows a Poisson distribution with  = 5.


#### Summary of Example
- `Exponential Distribution`: Provides insight into the time between successive events (phone calls). In our example, it models how long you wait between receiving calls, assuming calls come in at an average rate of 5 per hour.

- `Poisson Distribution`: Provides insight into the number of events within a fixed interval. In our example, it models how many calls you receive in an hour, assuming an average rate of 5 calls per hour.

#### Determining if a Process Follows a Poisson Distribution
--------------------------------------

##### Characteristics of a Poisson Process

1. **Events Occur Independently:**
   - The occurrence of one event does not affect the probability of another event occurring. Each event is independent of others.

2. **Constant Average Rate:**
   - Events occur at a constant average rate \( \lambda \) (the rate parameter). The rate does not change over time or space.

3. **Events Occur in Non-overlapping Intervals:**
   - The number of events in non-overlapping intervals of time or space is independent of each other. For instance, the number of customers arriving in one hour is independent of the number arriving in the next hour.

4. **Rare Events Over Small Intervals:**
   - The probability of more than one event occurring in a very small interval is negligible. This means the process is "memoryless," and the probability of an event occurring in a short time interval is proportional to the length of that interval.

##### Steps to Determine if a Process Follows a Poisson Distribution

1. **Check Independence:**
   - Verify that the events occur independently. For instance, in a call center, check if the arrival of one call is not dependent on the arrival of another.

2. **Estimate the Rate ():**
   - Calculate the average rate of occurrences (). For instance, count the number of events in several time intervals and find the average.

3. **Verify Constant Rate:**
   - Ensure that the rate of occurrences remains constant over time. Analyze whether the rate changes in different periods or conditions.

4. **Compare to Poisson Distribution:**
   - **Empirical Test:** Collect data on the number of events in fixed intervals. Compare this data to the expected frequencies given by the Poisson distribution with the estimated rate \( \lambda \).
   - **Goodness-of-Fit Test:** Perform statistical tests (e.g., Chi-square test) to compare observed frequencies with expected frequencies from a Poisson distribution.

5. **Check for Rare Events:**
   - Confirm that the probability of more than one event occurring in a very small time interval is negligible.

##### Practical Example: Call Center Arrivals

1. **Collect Data:**
   - Record the number of calls received per hour over several days.

2. **Calculate Rate ():**
   - Compute the average number of calls per hour.

3. **Perform Tests:**
   - Use the data to compare observed counts of calls with the expected counts from a Poisson distribution with the calculated rate.

4. **Statistical Testing:**
   - Conduct a goodness-of-fit test to see if the observed distribution of call counts matches the Poisson distribution.


#### left skew

```python
data_left_skewed = np.random.normal(loc=10, scale=5, size=1000)
data_left_skewed = np.concatenate([data_left_skewed, np.random.normal(loc=20, scale=1, size=200)])

plt.figure(figsize=(5, 5))
sns.histplot(data_left_skewed, kde=True)
plt.title('Left-Skewed Distribution - 1200 Samples, Range: {:.2f} to {:.2f}'.format(data_left_skewed.min(), data_left_skewed.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
data_right_skewed = np.random.exponential(scale=2, size=1000)
data_right_skewed = np.concatenate([data_right_skewed, np.random.exponential(scale=8, size=200)])

plt.figure(figsize=(5, 5))
sns.histplot(data_right_skewed, kde=True)
plt.title('Right-Skewed Distribution - 1200 Samples, Range: {:.2f} to {:.2f}'.format(data_right_skewed.min(), data_right_skewed.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
