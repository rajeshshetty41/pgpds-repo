```python
from PIL import Image, ImageEnhance, ImageOps
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load an image
def load_image(image_path):
    return Image.open(image_path)

# Save an image
def save_image(image, save_path):
    image.save(save_path)
```

```python
# Rotation
def rotate_image(image, angle):
    return image.rotate(angle)

# Translation (manual implementation)
def translate_image(image, x_offset, y_offset):
    width, height = image.size
    # Create a transformation matrix for translation
    matrix = (1, 0, x_offset, 0, 1, y_offset)
    return image.transform((width, height), Image.AFFINE, matrix, fillcolor=(255, 255, 255))

# Scaling
def scale_image(image, scale_factor):
    width, height = image.size
    new_size = (int(width * scale_factor), int(height * scale_factor))
    return image.resize(new_size)

# Flipping
def flip_image(image, flip_type='horizontal'):
    if flip_type == 'horizontal':
        return ImageOps.mirror(image)
    elif flip_type == 'vertical':
        return ImageOps.flip(image)

# Color Jittering
def jitter_image(image, brightness=1.0, contrast=1.0, color=1.0, sharpness=1.0):
    enhancer = ImageEnhance.Brightness(image)
    image = enhancer.enhance(brightness)
    enhancer = ImageEnhance.Contrast(image)
    image = enhancer.enhance(contrast)
    enhancer = ImageEnhance.Color(image)
    image = enhancer.enhance(color)
    enhancer = ImageEnhance.Sharpness(image)
    image = enhancer.enhance(sharpness)
    return image
```

```python
# Display images
def display_images(original, rotated, translated, scaled, flipped, jittered):
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes[0, 0].imshow(original)
    axes[0, 0].set_title('Original Image')
    axes[0, 0].axis('off')
    
    axes[0, 1].imshow(rotated)
    axes[0, 1].set_title('Rotated Image')
    axes[0, 1].axis('off')
    
    axes[0, 2].imshow(translated)
    axes[0, 2].set_title('Translated Image')
    axes[0, 2].axis('off')
    
    axes[1, 0].imshow(scaled)
    axes[1, 0].set_title('Scaled Image')
    axes[1, 0].axis('off')
    
    axes[1, 1].imshow(flipped)
    axes[1, 1].set_title('Flipped Image')
    axes[1, 1].axis('off')
    
    axes[1, 2].imshow(jittered)
    axes[1, 2].set_title('Jittered Image')
    axes[1, 2].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Main script
image_path = r'C:\Users\bhupe\Desktop\aria.png'
```

```python
# Load image
original_image = load_image(image_path)
```

```python
# Apply augmentations
rotated_image = rotate_image(original_image, 45)
translated_image = translate_image(original_image, 20, 30)
scaled_image = scale_image(original_image, 3.5)
flipped_image = flip_image(original_image, 'horizontal')
jittered_image = jitter_image(original_image, brightness=1.2, contrast=1.5, color=1.1, sharpness=1.3)
```

```python
# Display images
display_images(original_image, rotated_image, translated_image, scaled_image, flipped_image, jittered_image)
```
