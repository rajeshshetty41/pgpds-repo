```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Customer data: age and income
customer_data = np.array([
    [25, 50000],
    [30, 75000],
    [35, 60000],
    [40, 80000],
    [45, 90000]
])
```

```python
# Select three customers
selected_customers = customer_data[:3]
```

```python
# Plot the customer data
plt.figure(figsize=(8, 6))

# Plot all customers' data
plt.scatter(customer_data[:, 0], customer_data[:, 1], color='blue', label='All Customers')

# Highlight the selected customers
plt.scatter(selected_customers[:, 0], selected_customers[:, 1], color='red', label='Selected Customers')

# Add labels for the selected customers
for i, (age, income) in enumerate(selected_customers):
    plt.annotate(f'Customer {i+1}\n({age}, {income})', (age, income), textcoords="offset points", xytext=(-15, 7), ha='center', fontsize=9, color='red')

# Plot vectors from origin to selected customers
origin = [0, 0]
vectors = selected_customers - origin

for i, (age, income) in enumerate(selected_customers):
    plt.quiver(*origin, vectors[i, 0], vectors[i, 1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
    plt.annotate(f'Vector {i+1}', (vectors[i, 0]/2, vectors[i, 1]/2), textcoords="offset points", xytext=(-10, 7), ha='center', fontsize=9, color='green')

# Add labels and legend
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Customer Information Data with Vectors')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

#### scalars

```python
# Define two scalar values
scalar1 = 5
scalar2 = 3
```

```python
# Plot a number line
plt.figure(figsize=(8, 1))
plt.axhline(y=0, color='black', linewidth=2)

# Mark the scalar values
plt.scatter(scalar1, 0, color='blue', label='Scalar 1')
plt.scatter(scalar2, 0, color='red', label='Scalar 2')

# Add labels for the scalar values
plt.annotate(f'Scalar 1: {scalar1}', (scalar1, 0), textcoords="offset points", xytext=(-20, 10), ha='center', fontsize=9, color='blue')
plt.annotate(f'Scalar 2: {scalar2}', (scalar2, 0), textcoords="offset points", xytext=(20, 10), ha='center', fontsize=9, color='red')

# Set plot limits and labels
plt.xlim(min(scalar1, scalar2) - 2, max(scalar1, scalar2) + 2)
plt.ylim(-5, 10)
plt.xticks([])  # Hide x-axis labels
plt.yticks([])  # Hide y-axis labels
plt.title('Scalar Values on Number Line')
plt.legend()

# Show the plot
plt.show()
```

E.g- when we multiply two records (vectors) -> scalar -> the strength of relation among the vectors
