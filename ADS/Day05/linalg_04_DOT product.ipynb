#### Ex 1 : Dot product

````python
import numpy as np
````

````python
# Define two vectors
vector_A = np.array([3, 4])
vector_B = np.array([1, 2])
````

````python
# Calculate the dot product
dot_product = np.dot(vector_A, vector_B)

print(f"The dot product of A and B is {dot_product}")
````

````python
# Calculate the magnitudes of the vectors
magnitude_A = np.linalg.norm(vector_A)
magnitude_B = np.linalg.norm(vector_B)
````

````python
# Calculate the angle between the vectors in degrees
angle_degrees = np.degrees(np.arccos(dot_product / (magnitude_A * magnitude_B)))

print(f"The angle between A and B is {angle_degrees} degrees")
````

# Intuition Behind the Dot Product

## What is the Dot Product?
The dot product (or scalar product) of two vectors is a single number obtained by multiplying corresponding elements of the vectors and summing up the results. It measures how much one vector extends in the direction of another.

## Intuitive Explanation:
- **Geometric Interpretation**: The dot product provides a measure of how much two vectors align with each other. If the vectors are pointing in the same direction, the dot product is large. If they are perpendicular (at a 90Â° angle), the dot product is zero. If they point in opposite directions, the dot product is negative.
- **Magnitude and Direction**: It combines the magnitudes of the vectors and the cosine of the angle between them. The formula is:
  $$[
  \text{Dot Product} = \| \mathbf{a} \| \cdot \| \mathbf{b} \| \cdot \cos(\theta)
  ]$$
  
  where \(\| \mathbf{a} \|\) and \(\| \mathbf{b} \|\) are the magnitudes of vectors \(\mathbf{a}\) and \(\mathbf{b}\), and \(\theta\) is the angle between them.
